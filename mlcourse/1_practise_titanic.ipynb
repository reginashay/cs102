```python
def age_category(age):
    '''
    < 30 -> 1
    >= 30, <55 -> 2
    >= 55 -> 3
    '''
    if age < 30:
        return 1
    elif age < 55:
        return 2
    else:
        return 3
```

```python
age_categories = [age_category(age) for age in data.Age]
```

```python
data['Age_category'] = age_categories
```

```python
data.head()
```

```python
data['Age_category'] = data['Age'].apply(age_category)
```

```python
def name(data):
    L = [i.split('. ')[1] for i in data[data['Sex'] == 'male']['Name']]
    for j in L:
        if ' ' in j:
            L.remove(j)
            L.append(j.split(' ')[0])
    return L
```
